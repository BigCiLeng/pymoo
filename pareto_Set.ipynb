```python
import numpy as np
```

```python
dataset_dir='/share/code/dataset/ship_design/'
table_vals = np.loadtxt(dataset_dir+"pareto_set.csv", delimiter=",")
index=[]
for i in table_vals:
    index.append(int(i[0]))
print(index)
```

```python
data=np.loadtxt(dataset_dir+'main_value.csv',delimiter=',')
ori_pareto_set=[]
for i in index:
    ori_pareto_set.append(data[i])
print(ori_pareto_set)
```

```python
import pandas as pd
pd.DataFrame(np.array(ori_pareto_set)).to_csv(dataset_dir+"ori_pareto_set.csv", mode="a", float_format="%.2f", index=False, header=0)
    
```
